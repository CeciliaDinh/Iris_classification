**DATA VISUALIZATION**

```python
plt.figure(figsize=(15,9))
sns.barplot(list(df.region_txt.value_counts().head(10).keys()),df.region_txt.value_counts().head(10),palette='husl')
plt.title('AFFECTED REGIONS ')
plt.xticks(rotation=80)
```

```python
pd.crosstab(df.iyear,df.region_txt).plot(kind='line',figsize=(20,9))
plt.title('Terrorist activity in several regions in 1965-2016')
plt.xlabel('Region')
plt.ylabel('Number of Attacks')
plt.xticks([1965,1970,1975,1980,1985,1990,1995,2000,2005,2010,2015,2016])
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(15,9))
sns.barplot(list(df.country_txt.value_counts().head(10).keys()),df.country_txt.value_counts().head(10),palette='Spectral')
plt.xlabel('COUNTRIES')
plt.ylabel('NUMBER OF TERRIOST ACTIVITY')
plt.title('AFFECTED COUNTRIES ')
plt.xticks(rotation=80)
```

```python
plt.figure(figsize=(15,9))
sns.barplot(list(df.city.value_counts().head(10).keys()),df.city.value_counts().head(10),palette='coolwarm')
plt.xlabel('CITIES')
plt.ylabel('NUMBER OF TERRIOST ACTIVITY')
plt.title('AFFECTED CITIES ')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(20,8))
sns.countplot('iyear',data=df,palette='Paired')
plt.title( "TERRITORY ACTIVITIES IN EACH YEAR")
plt.xlabel('YEAR')
plt.ylabel('CASES')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(20,8))
sns.countplot('attacktype1_txt',data=df,palette='rocket')
plt.xticks(rotation=90)
plt.title("METHOD OF ATTACKS")
plt.xlabel('ATTACKS')
plt.ylabel('CASES')
plt.show()
```

```python
plt.figure(figsize=(20,8))
sns.countplot('targtype1_txt',data=df,palette='tab20')
plt.xticks(rotation=90)
plt.title("ATTACK'S TARGETS")
plt.xlabel('TARGETS')
plt.ylabel('CASES')
plt.show()
```

```python


plt.figure(figsize=(20,8))
sns.countplot('gname',data=df.head(10),palette='gist_heat_r')
plt.xticks(rotation=90)
plt.title("TERRITORY ACTIVITIES GROUP")
plt.xlabel('GROUP')
plt.ylabel('Counts')
plt.show()
```
